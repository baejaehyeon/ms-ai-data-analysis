# 파이썬 데이터 분석

> 데이터 분석 5단계
- 문제 정의 : 현재 상황에 따른 문제점을 파악하고, 미래 개선될 점에 대해서 분석 및 정의
- 데이터 수집 : 문제 정의에 따라서 분석에 필요한 데이터 선택 및 수집
- 데이터 가공 : 확보한 데이터 관찰 및 기초 데이터 분석, 분석에 필요한 데이터 컬럼 추가 및 가공
- 데이터 분석 : 문제 정의를 반영한 데이터 분석과, 연관될 수 있는 다른 데이터 수집 및 연결
- 시각화 및 탐색 : 데이터 분석 결과를 반영할 수 있는 시각화 및 문제 정의에 맞춘 해결방안 도출
