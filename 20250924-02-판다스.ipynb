````python
### 8.7 데이터에 리터럴 값 결합

"""
- 데이터프레임[컬럼].apply(함수) : 각 데이터에 함수가 리턴한 값으로 채움. 데이터 편집 시 용이
"""


#```

import pandas as pd

df = pd.read_csv('heroes.csv')

# 함수
def add_cm(hei) :
    # '175' + 'cm'
    return str(hei) + 'cm' 

def remove_cm(hei) :
    # '175cm' => '175'
    return hei.replace('cm', '')


df['키'] = df['키'].apply(add_cm)
df['키'] = df['키'].apply(remove_cm)


df

#```
````
